# Model

```python
# ollama needs to be installed
# https://ollama.com/
# https://github.com/ollama/ollama/tree/main/docs

from langchain_ollama.llms import OllamaLLM

model = OllamaLLM(model="llama3.2")

questions = ["Quais so os ingredientes do Condotril?",
"Quais so os ingredientes do Duobiotic?",
"Quais so os ingredientes do Neurofil?",
"Quais so os benefcios do Condotril?",
"Quais so os benefcios do Duobiotic?",
"Quais so os benefcios do Neurofil?",
"Quais so os efeitos do Condotril?",
"Quais so os efeitos do Duobiotic?",
"Quais so os efeitos do Nueurofil?",
"Qual  o processo de toma do Condotril?",
"Qual  o processo de toma do Duobiotic?",
"Qual  o processo de toma do Neurofil?",
"Quais so as condies de armazenamento do Condotril?",
"Quais so as condies de armazenamento do Duobiotic?",
"Quais so as condies de armazenamento do Neurofil?",
"Em que casos no se pode tomar Condotril?",
"Em que casos no se pode tomar Duobiotic?",
"Em que casos no se pode tomar Neurofil?"]
```

## Variants Generation

```python
from langchain_core.messages import SystemMessage

def generate_variants(model, questions, n=3):
    variants = {}
    for question in questions:
        system_prompt =  """Your role is to rephrase this question in different ways, keeping the meaning: {question}""".format(question=question)

        response = model.generate(prompts=[system_prompt], num_return_sequences=n)
        variants[question] = [variant for variant in response.generations]
    return variants
        
```

```python
question_variants = generate_variants(model, questions, n=3)
```

```python
for original, variants in question_variants.items():
    print(f"Original: {original}")
    for i, variant in enumerate(variants, 1):
        print(f"Variant {i}: {variant}")
    print()
```

## Display das variantes a cada pergunta

```python
for original, variants in question_variants.items():
    print(f"Original: {original}")
    for i, variant in enumerate(variants, 1):
        responses = variant[0].text.split("\n- ")
        responses = [resp.strip("- ") for resp in responses[1:] if resp.strip()]
        for i, resp in enumerate(responses, start=1):
            print(f"Variante {i}: {resp}")
    print()
```
